```python
import pandas as pd
import requests
from bs4 import BeautifulSoup #web scrapping library (bs4 = beautiful soup 4)
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://collegedunia.com/india-colleges', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, "lxml")
```

```python
soup.find_all('h3', class_= "jsx-3609176813 font-weight-medium text-lg mb-0")[4].text
```

```python
college = soup.find_all('div', class_='jsx-3609176813 clg-column d-flex align-items-start')
course_fees = soup.find_all('a', class_='jsx-3609176813 pointer underline-on-hover')
placement = soup.find_all('td', class_='jsx-3609176813 col-placement')
review = soup.find_all('span', class_='jsx-3609176813 lr-key text-lg text-primary d-block font-weight-medium mb-1')
rank = soup.find_all('div', class_='jsx-1890585645 modal-table f-12')
```

```python
name=[]
city=[]
state=[]
fees=[]
course=[]
average_ctc=[]
highest_ctc=[]
reviews=[]
link=[]


for i in college:
    name.append(i.find('h3', class_= "jsx-3609176813 font-weight-medium text-lg mb-0").text)
    city.append(i.find('span', class_="jsx-3609176813 pr-1 location").text.split(',')[0].strip())
    state.append(i.find('span', class_="jsx-3609176813 pr-1 location").text.split(',')[1].strip())
    link.append(i.find('a' , href = True)['href'])

for i in course_fees:
    fees.append(i.find('span', class_="jsx-3609176813 text-lg text-green d-block font-weight-bold mb-1").text.replace('₹', '').strip())
    course.append(i.find('span', class_="jsx-3609176813 fee-shorm-form").text)
    
if placement:
    for i in range(len(placement)):
        ctc_temp = placement[i].text.replace('₹\xa0', '').replace('Compare Placement' , '')
        average_ctc.append(ctc_temp.split('Average Package')[0].replace('Highest Package', '').strip())
        highest_ctc.append(ctc_temp.split("Highest Package")[0].split('Average Package')[-1].strip())

else:
    average_ctc.append('NA')
    highest_ctc.append("NA")

for i in range(len(review)):
    reviews.append(review[i].text.replace('/ 5 ', '').strip())


```

```python
final = {'College Name':name, 'City':city, 'State':state, 'Course':course, "Fees":fees, "Average Package":average_ctc, 'Highest Package':highest_ctc,'Reviews':reviews, 'College Website Link': link}
```
